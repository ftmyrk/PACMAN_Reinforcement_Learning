```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np

# Load the .mat file
mat = sio.loadmat('training_data.mat')

# Extract episode rewards
rewards = mat['episode_rewards'][0]

# Calculate the mean reward
mean_reward = np.mean(rewards)

# Plot the rewards
plt.figure(figsize=(12, 6))
plt.plot(rewards, label='Episode Rewards')
plt.axhline(y=mean_reward, color='r', linestyle='--', label='Mean Reward')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Episode Rewards and Mean Reward')
plt.legend()
plt.show()
```

```python
total_mean = []
hundred_mean = []
total_reward = 0
hundreds = 0
for i in range(len(rewards)):
  total_reward += rewards[i]
  hundreds += rewards[i]
  total_mean.append(total_reward / (i + 1))
  if (i + 1) % 100 == 0:
    hundred_mean.append(hundreds / 100)
    hundreds = 0

# Plot the rewards
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(total_mean, label='Mean')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Mean for Every Episode')

plt.subplot(1, 2, 2)
plt.plot(hundred_mean, label='Hundred Mean')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Mean Reward for Every 100 Episode')
plt.show()
```
